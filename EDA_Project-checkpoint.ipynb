```python
df['rate'].unique() #To abstract the unique values of the feature-rate, to find out if string and - values are present
```

```python
#writing a code in order to replace 'New', '-', values with nan(as we can replace missing values), and removing the'/5' from rate
def handlerate(value):
    if value=='NEW' or value =='-':
        return np.nan
    else:
        value=str(value).split('/')
        value = value[0]
        return float(value)
df['rate']=df['rate'].apply(handlerate)
df['rate'].head()
```

```python
#Checking the total number of nan values in rate column
df.rate.isnull().sum()
```

```python
#replacing the missing values in the rate column with the mean values
df['rate'].fillna(df['rate'].mean(),inplace =True)
df['rate'].isnull().sum()
```

Cleaning the 'rest_type' column

```python
#To count the total number of each of the categories present
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000]
rest_types_lessthan1000
```

As can be seen above, there are many restaurants that has only a few categories namely, bar, lounge, dessert parlour and so on. So what we can do is, we shall cluster all those as "others" category.

```python
#Making rest types less than 1000 in frequency and renaming them as others:
def handle_rest_type(value):
    if (value in rest_types_lessthan1000):
        return 'others'
    else:
        return value
df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location =  df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]
def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value
df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```
